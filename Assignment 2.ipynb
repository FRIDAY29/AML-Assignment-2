# Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Initializing LabelEncoder
le = LabelEncoder()

# Encoding the 'gender' column
df['gender'] = le.fit_transform(df['gender'])

# Encoding the 'ssc_b' column
df['ssc_b'] = le.fit_transform(df['ssc_b'])

# Encoding the 'hsc_b' column
df['hsc_b'] = le.fit_transform(df['hsc_b'])

# Encoding the 'hsc_s' column
df['hsc_s'] = le.fit_transform(df['hsc_s'])

# Encoding the 'degree_t' column
df['degree_t'] = le.fit_transform(df['degree_t'])

# Encoding the 'specialisation' column
df['specialisation'] = le.fit_transform(df['specialisation'])

# Encoding the 'status' column (target variable)
df['status'] = le.fit_transform(df['status'])

# Encoding the 'workex' column
df['workex'] = le.fit_transform(df['workex'])

# Filling missing values in the 'salary' column with the mean salary
df['salary'] = df['salary'].fillna(df['salary'].mean())
```

```python
print(df)
```

# Modeling the data

```python
X = df.drop('status',axis=1)
y = df['status']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=7)
```

```python
print("Input Training:",X_train.shape)
print("Input Test:",X_test.shape)
print("Output Training:",y_train.shape)
print("Output Test:",y_test.shape)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

77% accuracy of logistic regression. hat's quite descent accuracy for our model.


```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n",confusion_matrix)
from sklearn.metrics import classification_report
print("Classification Report:\n",classification_report(y_test, y_pred))
```

The Confusion matrix result is telling us that we have 10+40 correct predictions and 5+10 incorrect predictions.
The Classification report reveals that we have 73% precision which means the accuracy that the model classifier not to label an instance positive that is actually negative and it is important to consider precision value because when you are hiring, you want to avoid Type I errors at all cost. They are culture killers.In hiring, a false positive is when you THINK an employee is a good fit, but in actuality theyâ€™re not.

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rt=RandomForestClassifier(n_estimators=100)
rt.fit(X_train,y_train)
y_pred=rt.predict(X_test)
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

We can see we have good accuracy for random forest model of aprox 92.3% which is good.

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
error_rate = []

for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
# Error rate vs K-value
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

Lets take K value 10.

```python
from sklearn.metrics import confusion_matrix
knn = KNeighborsClassifier(n_neighbors=10, metric='euclidean')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
confusion_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n",confusion_matrix)
print("Classification Report:\n",classification_report(y_test, y_pred))
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

For K value 10 our accuracy is  97%.
